Importing all the dependencis.

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

Data Collection and Data Preprocessing

```python
data = pd.read_csv('/newss.csv')
```

```python
X = data[['title', 'text', 'subject', 'date','label']]
```

```python
# Counting the missing Values in dataset.
X.isnull().sum()
```

```python
# Filling all the null values with empty string.
X = X.fillna('')
```

```python
A = X.drop(columns='label', axis=1)
B = X['label']
```

```python
# Stemming the data now. Its basically root form of the word. eg. running -> run.
port_stem = PorterStemmer()
```

```python
#def stemming(text):
 # stemmed_text = re.sub('[^a-zA-Z]', ' ', text)
  #stemmed_text = stemmed_text.lower()
  #stemmed_text = stemmed_text.split()
  #stemmed_text = [ port_stem.stem(word) for word in stemmed_text if not word in stopwords.words('english')]
  #stemmed_text = ' '.join(stemmed_text)
  #return stemmed_text
```

```python
def stemming(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    words = text.lower().split()
    return ' '.join(
        port_stem.stem(word) for word in words if word not in stop_words
    )
```

```python
A['text'] = A['text'].apply(stemming)
```

```python
print(A['text'])
```

```python
# Seperating data and labels.
C = A['text'].values
D = X['label'].values
```

```python
print(C)
```

```python
print(D)
```

```python
C = A['text'].values
vectorizer = TfidfVectorizer()
vectorizer.fit(C)
C = vectorizer.transform(C)
```

```python
print(C)
```

```python
C.shape
```

```python
# Splitting training and testing data
C_train, C_test, D_train, D_test = train_test_split(C, D, test_size=0.3, stratify = D, random_state=42)
```

Training the model: Logistic Regression.

```python
model = LogisticRegression()
```

```python
model.fit(C_train, D_train)
```

Evaluation: Accuracy Score.

```python
# Accuracy score on training data
C_train_prediction = model.predict(C_train)
train_accuracy = accuracy_score(C_train_prediction, D_train)
print('Accuracy Score is : ', train_accuracy)
```

```python
# Accuracy score on testing data
C_test_prediction = model.predict(C_test)
test_accuracy = accuracy_score(C_test_prediction, D_test)
print('Accuracy Score is : ', test_accuracy)
```

```python
# Making a Predictive System.
C_new = C_test[4]
prediction = model.predict(C_new)
print(prediction)

if(prediction[0] == 0):
    print('The news is Real')
else:
    print('The news is Fake')
```
